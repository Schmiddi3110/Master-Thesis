```python
import pandas as pd
import matplotlib.pyplot as plt
datasets = ["SpokenArabicDigits", "PenDigits", "WalkingSittingStanding"]
```

# Analysis of Substitute Model Parameters Count

```python
  
# Loop through datasets
for dataset in datasets:
    # Load CSV file
    file_path = f"TimeSeriesModels/{dataset}/Results/BaseAttacks/Base_Attack.csv"
    df = pd.read_csv(file_path)

    # Extract model endings
    df['Model Ending'] = df['Model Pairing'].str.split('-').str[1]  # Get the part after the dash

    # Group by model ending and calculate the average accuracy gain per parameter count, over stealing dataset size
    df['Accuracy per Param'] = df['Accuracy'] / df['Parameter Count']  # Element-wise division first

    grouped_df = df.groupby(['Model Ending', 'Stealing Dataset Size']).agg(
        avg_accuracy_gain_per_param_count=('Accuracy per Param', 'mean')).reset_index()


    # Plotting
    plt.figure(figsize=(12, 8))

    # Create plots for each model ending
    for model_ending in grouped_df['Model Ending'].unique():
        model_data = grouped_df[grouped_df['Model Ending'] == model_ending]
        plt.plot(model_data['Stealing Dataset Size'], model_data['avg_accuracy_gain_per_param_count'], label=model_ending)

    plt.title('Average Accuracy per Parameter Count Over Stealing Dataset Size')
    plt.xlabel('Stealing Dataset Size')
    plt.ylabel('Average Accuracy per Parameter Count')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Analysis of Model Architecture

```python

# List of dataset names
datasets = ["PenDigits", "WalkingSittingStanding", "SpokenArabicDigits"]

# Loop through datasets
for dataset in datasets:
    # Load CSV file
    file_path = f"TimeSeriesModels/{dataset}/Results/BaseAttacks/Base_Attack.csv"
    df = pd.read_csv(file_path)

    # Extract the original model from the 'Model Pairing' column (before the dash)
    df['Original Model'] = df['Model Pairing'].str.split('-').str[1]

    # Group by original model and stealing dataset size to get average attack duration
    duration_grouped_df = df.groupby(['Original Model', 'Stealing Dataset Size']).agg(
        avg_attack_duration=('Accuracy', 'mean')).reset_index()
    
    # Plotting attack duration
    plt.figure(figsize=(12, 8))
    for model in duration_grouped_df['Original Model'].unique():
        model_data = duration_grouped_df[duration_grouped_df['Original Model'] == model]
        plt.plot(model_data['Stealing Dataset Size'], model_data['avg_attack_duration'], label=model)
    
    plt.title(f'Average Substitute Model Accuracy Over Stealing Dataset Size on {dataset}')
    plt.xlabel('Stealing Dataset Size')
    plt.ylabel('Average Substitute Model Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

colors = ["blue", "green", "red"]  # Assigning colors for datasets

# Define unique markers for each attack method
markers = ["o", "s", "^", "D", "P", "*", "X"]  # Add more if needed

plt.figure(figsize=(12, 8))

# Dictionary to store markers for each attack method
method_markers = {}

# Loop through datasets
for dataset, color in zip(datasets, colors):
    # Load CSV file
    file_path = f"TimeSeriesModels/{dataset}/Results/BaseAttacks/Base_Attack.csv"
    df = pd.read_csv(file_path)

    # Group by Method Name and Stealing Dataset Size to get average attack accuracy
    grouped_df = df.groupby(['Method Name', 'Stealing Dataset Size']).agg(
        avg_attack_sub_Acc=('Accuracy', 'mean')).reset_index()

    # Loop through methods
    for method in grouped_df['Method Name'].unique():
        method_data = grouped_df[grouped_df['Method Name'] == method]

        # Assign a unique marker for each method
        if method not in method_markers:
            method_markers[method] = markers[len(method_markers) % len(markers)]

        plt.plot(method_data['Stealing Dataset Size'], method_data['avg_attack_sub_Acc'],
                 label=f"{method} ({dataset})", linestyle='-', marker=method_markers[method], color=color)

# Plot settings
plt.title('Average Attack Accuracy Over Stealing Dataset Size (All Datasets)')
plt.xlabel('Stealing Dataset Size')
plt.ylabel('Average Attack Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
